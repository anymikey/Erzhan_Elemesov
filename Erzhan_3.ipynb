### Create objects to train and test the model

```python
#Create a feature object from the columns
X_features = df_ad_budget_data_ind[['TV Ad Budget ($)', 'Radio Ad Budget ($)', 'Newspaper Ad Budget ($)']]
```

```python
#View the feature object
X_features.head()
```

```python
#Create a target object
y_target = df_ad_budget_data_ind[['Sales ($)']]
y_target.head()
```

```python
#Verify if all observations have been captured in the feature object
X_features.shape
```

```python
#Verify if all observations have been captured in the target object
y_target.shape
```

### Split the original dataset into training and testing datasets for the model

```python
#Import Scikit Learn function
from sklearn.model_selection import train_test_split
```

```python
#Split the dataset by 75:25
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, random_state=27)
```

```python
#Verify if the training and testing dataset have a correct split
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Create a model to predict the sales outcome

```python
#Create a linear regression model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
#sales -> constant + coef1*(TV Budget) + coef2*(Radio budget) + coef3*(Newspaper budget)
#print coefficients
print(linreg.intercept_)
print(linreg.coef_)
```

```python
#sales -> 2.79916791 + 0.04648022*(TV Budget) + 0.1845547*(Radio budget) + 0.00286741*(Newspaper budget)
#Predict the outcome for the testing dataset
y_pred = linreg.predict(X_test)
y_pred
```
